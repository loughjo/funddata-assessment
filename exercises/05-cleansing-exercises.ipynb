Write a Python function to remove all non-alphanumeric characters from a string.
***

Non-alphanumeric characters are characters that are not numbers (0-9) or alphabetic characters. 
So Alphanumeric characters are A to Z, a to z and 0 to 9. I assume spaces or underscores are not considered.

```python
# Import the re module which is used to work with regular expressions.
import re
```

The *removeNonAlphaNumeric* functions takes in a string and removes all non alphanumeric characters. 
The only characters it accepts are letters of the alphabet lower and uppercase a-z and A-Z. Numbers are also accepted 0-9.

The function *removeNonAlphaNumeric* uses the substitution function re.sub() to search for non-alphanumeric characters and
remove them.

```python
def removeNonAlphaNumeric(string):
   #[^\w] removes all characters apart from a-zA-Z0-9_ and [_] removes underscore '_'
    strip = re.sub(r'[^\w]|[_]',  '', string)
    return strip
```

```python
# This is the string we use to test the function
string = 'The big pig!    _          flew over the house? @&#>! pat% 12 g712 tftftftftf football(*702j>). The BIG Brown fox &* (-+ = {}[] Christmas)'
```

```python
# we pass the string variable through the function removeNonAlphaNumeric to remove all alpha numeric characters including
# whitespace characters.
removeNonAlphaNumeric(string)
```

## Capitizing first Letter in species using for loop

```python
url = r'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'

iris = [line.decode('utf-8').strip() for line in urllib.request.urlopen(url)]

iris
```

Before reading the exercise correctly I used the code below to Capitalize the first letter in each species. 
Just including it in the notebook.  

```python
for i in range(len(iris)-1):
    # iterate over each row and assign each iris index in turn to astring variable
    astring  = iris[i]
    # We split the string in two and its in a list e.g ['5.1,3.5,1.4,0.2,Iris', 'Setosa']
    newString = astring.split('-')
    # We capatize the first Letter in the species which is in index 1 in the list and 
    # concatenate string in index 0 + '-' + string in index 1 which is the species with capital letter. 
    word  = newString[0] + '-' + newString[1].capitalize()
    # we assign the updated string variable 'word' to its iris index
    iris[i] = word
    
# display the contents of the Iris list which show the first letter in Species capitalized.
iris
```
