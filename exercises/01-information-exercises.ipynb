```python
# Make HTTP requests for internet reqources.
import urllib.request

# The URL of a text version of Alice in Wonderland.
book_url = 'https://www.gutenberg.org/files/11/11-0.txt'

# Get the book.
book = list(urllib.request.urlopen(book_url))

# Decode the lines and strip line endings.
book = [line.decode('utf-8-sig').strip() for line in book]

# Get a sample paragraph - I looked for this by hand.
paragraph = ' '.join(book[58:63])

# Show the paragraph.
print(paragraph)
```

```python
# Let's lower-case it.
alice = paragraph.lower()

# All letters and a space.
chars = 'abcdefghijklmnopqrstuvwxyz '

# And strip anything that is not a letter or space.
alice = ''.join([c for c in alice if c in chars])

# Show the paragraph now.
print(alice)
```

```python
# For doing (pseudo-)random things in Python.
import random

# Print a random character, for example.
print(random.choice(chars))
```

```python
# Get the length of alice.
N = len(alice)

# Generate N random characters from chars.
gener = random.choices(chars, k=N)

# Join them together in a string.
gener = ''.join(gener)

# Print.
print(gener)
```

```python
# Get the whole book in one big string.
sbook = ''.join(book[26:]).lower()

# Create the weights - count the occurences of each character in the whole book.
weights = [sbook.count(c) for c in chars]

# Show the weights.
weights
```

```python
# Generate a string using those weights.
wgenr = random.choices(chars, weights=weights, k=N)

# Join them together in a string.
wgenr = ''.join(wgenr)

# Print.
print(wgenr)
```

## Exercise 1 - *Part One* - Code Changes for Previous Two Characters
***

```python
# Create the weights call it threeghts so we have a map of weights for 'aaa', 'aab', 'aac'.........
threeghts = {z: {c: {d: sbook.count(z + c + d) for d in chars} for c in chars} for z in chars}
# Show the weights.
threeghts
#twoghts
```

```python
# Loop through our character set.
for i in range(len(chars)):
    # Print the character and how many times it appears in Alice in Wonderland.
    print(f'{chars[i]}: {weights[i]}')
```

## Exercise 1 - Part Two - Code Changes for Previous Two Characters

```python
# Start with i and space.
pairs = 'i '

# Do the following N-1 times.
for i in range(1, N):
    # Get the last 2 characters which are required to find the weight
    last2chars = (pairs[-2:])
    # Dictionary within a dictionary so first get all weights starting with 'i'
    secondLastChar = threeghts[last2chars[0]]
    # Get all weights for 'i ' 
    wt = secondLastChar[last2chars[1]] 
    # Turn wt into a list, ordered by chars.
    wt = [wt[c] for c in chars]
    # Randomly pick the next character using those weights.
    nextc = random.choices(chars, weights=wt, k=1)[0]
    # Append the character to twos.
    pairs = pairs + nextc
```

#### Explain why the log of zero is undefined.
***

The log function is undefined when the argument is 0 or negative.

**The variables in the equations:**<br/>
B = the base of the logarithm<br/>
E = the exponent (or output) of the logarithm<br/>
N = the number (or argument) of the logarithm<br/>

Here is an equation with the variables

$log_{B}(N) = E$ 

Using equations we will see why the log of zero is undefined. 
Below we have log 0 to the base 2 equals the exponent - E

$log_{2}(0) = E$ 

We now convert from log form to exponential we get the following :

$2^{E} = 0$

We get 2 to the power of the exponent - E equals 0.

Now if you substitute E for any number you can't make the equation equal to zero. Therefore the log of zero is undefined.
